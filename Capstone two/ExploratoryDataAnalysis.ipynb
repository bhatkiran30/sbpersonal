```python
# First we will compare sex and chestpain column and see how they are diversified based on target
_ = sns.swarmplot(x='sex',y='cp',hue='target',data=df,dodge=True)
```

```python
# First we will compare sex ad chestpain column and see how they are diversified based on target
_ = sns.violinplot(x='sex',y='cp',hue='target',data=df,dodge=True)
```

```python
# Now we will combine both swarm  plot and violin plot to get a more in depth idea
_ = plt.figure(figsize=(10,10))
_ = sns.violinplot(x='sex',y='cp',hue='target',data=df,dodge=True)
_ = sns.swarmplot(x='sex',y='cp',hue='target',data=df,dodge=True)
```

As we can see combining both the plots give us a more detailed description of how the data is diversified

```python
# Now we will see how thal disorder is reflected based on the sex of each data and how they are grouped based on our target variable
_ = plt.ylim([-1,4])
_ = sns.violinplot(x='sex',y='thal',hue='target',data=df,dodge=True)
_ = sns.swarmplot(x='sex',y='thal',hue='target',data=df,dodge=True)
```

From the above plot we can come to the following conclusions
* People with Thalassemia value of 2 i.e fixed defect are more prone to heart attack
* People with recoverable Thalassemia defect i.e 3 are comparitively less prone to heart attack
* Male patients have a higher count of thal order 3

```python
# Now we will plot chest pain and age to see how they are related and how they are grouped based on our target column
_ = sns.jointplot(x='age',y='cp',data=df,kind='kde')
```

```python
# Now we will plot a heat map to see how each feature is correlated with one another
plt.figure(figsize=(40,20))
sns.heatmap(df.corr(),annot=True,annot_kws={"size": 26})
```
